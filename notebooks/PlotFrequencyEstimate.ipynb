```python
def read(dirname):
    result_paths = glob.glob(os.path.join(dirname, '*tar'))
    results = []
    for p in result_paths:
        result = torch.load(p).get('result')
        results.append(result)

    results = sorted(results, key=lambda x: x.usr_configs.signal.freqs[0])
    return results
```

```python
def get_freq_error(dirname):
    results = read(dirname)
    N = results[0].usr_configs.signal.block_size
    fs = results[0].usr_configs.signal.fs
    num_pos = results[0].usr_configs.signal.num_pos_decision
    sMSEs = []
    true_f0s = []
    for i in range(len(results)):
        f0s = fs*results[i].scores.indices[0:num_pos]/N
        true_f0 = results[i].usr_configs.signal.freqs[0]
        
        amps = 2*np.sqrt(results[i].scores.scores[0:num_pos])
        true_amps0 = results[i].usr_configs.signal.amps[0]
        
        MSEf = np.mean((f0s-true_f0)**2)
        MSEA = np.mean((amps-true_amps0)**2)
        sMSE = np.sqrt(MSEA)

        sMSEs.append(sMSE)
        true_f0s.append(true_f0)
    return true_f0s, sMSEs
```

```python
true_f0s_ml1,sMSEs_ml1=get_freq_error(ml1)
true_f0s_ml10,sMSEs_ml10=get_freq_error(ml10)
true_f0s_switched1,sMSEs_switched1 = get_freq_error(switched1)
true_f0s_switched10,sMSEs_switched10 = get_freq_error(switched10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(true_f0s_ml1,np.array(sMSEs_ml1),marker='o',markersize=5,label='DFT - Option I (L=1)')
plt.plot(true_f0s_ml10,np.array(sMSEs_ml10),marker='o',markersize=5,label='DFT - Option I (L=10)')
plt.plot(true_f0s_switched1,np.array(sMSEs_switched1),marker='o',markersize=5,label='DFT - Option II (L=1)')
plt.plot(true_f0s_switched10,np.array(sMSEs_switched10),marker='o',markersize=5,label='DFT - Option II (L=10)')
plt.grid()
plt.xlabel('true '+'$f_o$'+'(Hz)',fontsize=15)
#plt.ylabel('$\sqrt{E[|f_o-\hat{f_o}|^2]}$'+'(Hz)',fontsize=15)
plt.ylabel('$\sqrt{E[|A-\hat{A}|^2]}$',fontsize=15)
plt.tick_params('both',labelsize=15)
plt.legend(fontsize=15)
plt.savefig('../plots/f.png')
```
